```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the differential equation for logistic growth
def logistic_growth(P, t, r, K):
    dPdt = r * P * (1 - P / K)
    return dPdt
def gompertz_growth (P, t, r, K):
    dPdt = r *np.log(K/P)
    return dPdt
def richards_growth(P,t,r,k,beta):
    dPdt= r*P*(1 - (P / K)**beta)
    return dPdt
def blumberg_growth(P,t,r,k,gamma)
    dPdt= r*P*(1 - (P / K)**gamma
# Parameters
r = 0.1  # Growth rate
K = 100  # Carrying capacity
P0 = 10  # Initial population
t = np.linspace(0, 200, 100)  # Time span for simulation
beta=2
gamma=1.5
# Solve the differential equation
P = odeint(logistic_growth, P0, t, args=(r, K))
Q = odeint (gompertz_growth, P0, t, args=(r, K))
R= odeint (richards_growth, P0, t, args=(r, K, beta))
S= odeint (blumberg_growth, P0, t, args=(r, K, gamma))


# Plot the results
plt.plot(t, P, 'tangerine', label='None')
plt.plot(t, Q, 'yellow', label='Weak')
plt.plot(t, R, 'salmon', label='Strong')
plt.plot(t, S, 'turquoise', label='Blumberg')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Regulated Growth Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
r = 0.1  # Intrinsic growth rate
K = 10  # Carrying capacity
L=3

# Define population range
N = np.linspace(0, K, 1000)

# Calculate per capita growth rate
per_capita_growth_rate = r * (1 - N / K)*N
per_capita_growth_rate2= r * (1 - (N / K))*(N**2)/5
per_capita_growth_rate3= r * (1 - (N / K))*(N/L-1)

# Plot
plt.plot(N, per_capita_growth_rate, color='yellow', label='No' )
plt.plot(N, per_capita_growth_rate2, color='salmon', label='Weak' )
plt.plot(N, per_capita_growth_rate3, color='#F08000', label='Strong' )

plt.title('Growth rates with Allee effects')
plt.xlabel('Population Size')
plt.ylabel('Growth Rate')
plt.ylim(-0.2,0.4,10)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Adding a horizontal line at y=0
plt.legend()
plt.grid(True)
plt.show()
```
