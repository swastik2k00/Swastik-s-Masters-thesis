```python
import numpy as np
import matplotlib.pyplot as plt



r1=0.5
r2=1
b1=2
b2=1
k1=2
k2=3
def f(Y, t, r1, r2, b1, b2,k1,k2):
    y1, y2 = Y
    return [r1*y1*(1-y1/k1)-b1*y1*y2, r2*y2*(1-y2/k2)-b2*y1*y2]

y1 = np.linspace(0, 10, 100)
y2 = np.linspace(0, 10, 100)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y],t, r1, r2, b1, b2,k1,k2)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]


plt.streamplot(Y1, Y2, u, v, color='#FF033E')
plt.axis('square')
plt.axis([0, 10, 0, 10])

plt.title("Principle of Competitive Exclusion")
plt.xlabel ("x")
plt.ylabel ("y")


plt.show()



```

```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

R1=3/2
k1=3
R2=5
k2=1
def f(x,R1,k1):
   return 0*x

x = np.linspace(-0, 10, 100)

plt.plot(x, f(x,R1,k1), color='blue')

def g(x,R1,k1):
   return R1*(1-x/k1)

x = np.linspace(-0, 10, 100)

plt.plot(x, g(x,R1,k1), color='red')

plt.axvline(x = 0, color = 'r')

def fprime(x, R2):
   return k2*(1-x/R2)

x = np.linspace(-0, 10, 100)

plt.plot(x, fprime(x,R2), color='blue')
plt.title("Nullclines for interspecies competition, no Allee effect")

plt.show()
```
